### XOR problem

```python
x = [True, True, False, False]
y = [True, False, True, False]
print(np.logical_xor(x, y))
```

```python
np.random.seed(0)
X = np.random.randn(200, 2)
Y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0)
Y = np.where(Y, 1, 0)
```

```python
x1 = X[:, 0]
x2 = X[:, 1]
```

```python
plt.scatter(x1, x2, c=Y)
plt.grid('on')
plt.axis('equal')
```

<center>$(x_1, x_2) \;\;\; \rightarrow \;\;\; \phi(x) = (x_1^2, \sqrt{2}x_1x_2, x_2^2)$</center>

```python
x1 = X[:, 0]**2
x2 = X[:, 1]**2
x3 = np.sqrt(2)*X[:, 0]*X[:, 1]
```

```python
#%matplotlib notebook  # 3d plot 움직일 수 있음
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1,x3,x2, c=Y)
```

## SVM Test #1

2D input data (XOR) & Linear kernel

```python
from sklearn.svm import SVC

model1 = SVC(kernel="linear")
model1.fit(X, Y)
```

```python
PlotSVM2D(X, Y, model1, "XOR classification result using SVC")
plt.show()
```

## SVM Test #2

3D input data (XOR) & Linear kernel

```python
X2 = np.vstack([x1**2, np.sqrt(2)*X[:, 0]*X[:, 1], x2**2]).T
```

```python
X2.shape
```

```python
model2 = SVC(kernel ='linear')
```

```python
model2.fit(X2, Y)
```

```python
PlotSVM3D(X2, Y, model=model2)
```

## SVM Test #3

2D input data (XOR) & RBF kernel

```python
model3 = SVC(kernel="rbf")
model3.fit(X, Y)
```

```python
PlotSVM2D(X, Y, model3, "XOR classification result using SVC")
plt.show()
```
