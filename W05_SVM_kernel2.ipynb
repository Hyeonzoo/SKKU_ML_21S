```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
from util.plot_util import PlotSVM2D
from util.plot_util import PlotSVM3D
```

### Data preparation

```python
from sklearn.datasets import make_circles
```

```python
X, Y = make_circles(n_samples=500, noise = 0.02)
```

```python
X.shape
```

```python
Y.shape
```

```python
Y[:10]
```

```python
x1 = X[:, 0]
x2 = X[:, 1]
```

```python
plt.scatter(x1, x2, c=Y)
plt.grid('on')
plt.axis('equal')
```

* Making another dimension

```python
x3 = x1**2 + x2**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, x3, c=Y)
```

## SVM Test #1
2D input data (circular) & Linear kernel

```python
from sklearn.svm import SVC
model1 = SVC(kernel ='linear')
```

```python
model1.fit(X, Y)
```

```python
PlotSVM2D(X, Y, model=model1, title='Linear SVM')
```

## SVM Test #2
3D input data (circular) & Linear kernel

```python
X2 = np.vstack((x1, x2, x3)).T
```

```python
X2.shape
```

```python
model2 = SVC(kernel ='linear')
```

```python
model2.fit(X2, Y)
```

```python
PlotSVM3D(X2, Y, model=model2)
```

## SVM Test #3
2D input data (circular) & RBF kernel

```python
model3 = SVC(kernel ='rbf')
```

```python
model3.fit(X, Y)
```

```python
PlotSVM2D(X, Y, model=model3, title='RBF SVM')
```
